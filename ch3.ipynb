```python
#환경설정
from mlwpy import *
%matplotlib inline
```

```python
iris = datasets.load_iris()
print(type(iris))
print(iris.keys())
iris_df = pd.DataFrame(iris.data,
                       columns=iris.feature_names)
iris_df['target'] = iris.target
display(pd.concat([iris_df.head(3),
                   iris_df.tail(3)]))
# pd.concat 는 DataFrame 합치는데 사용하는 판다스
```

```python
sns.pairplot(iris_df, hue='target', size= 1.5, palette='husl');
# seaborn 에서 제공하는 sns.pairplot(데이터) 는
# 각 column 별 데이터에 대한 상관관계나 분류적 특징을 명령어 한줄로 확인가능
# hue는 범례 나타내는거!
```

```python
print('targets: {0}'.format(iris.target_names),
      iris.target_names[0], sep='\n')
```
